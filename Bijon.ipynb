```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_line(x1, y1, x2, y2):
    
    is_steep = abs(y2 - y1) > abs(x2 - x1)
    
    if is_steep:
        x1, y1 = y1, x1
        x2, y2 = y2, x2
        
    sx = 1 if x1 < x2 else -1
    sy = 1 if y1 < y2 else -1
    
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    
    p = 2 * dy - dx
    
    x_values = []
    y_values = []
    x = x1
    y = y1
    
    while True:
        if is_steep:
            x_values.append(y)
            y_values.append(x)
        else:
            x_values.append(x)
            y_values.append(y)
        
        if x == x2 and y == y2:
            break
            
        if p < 0:
            p += 2 * dy
        else:
            y += sy
            p += 2 * dy - 2 * dx
        
        x += sx
        
    return x_values, y_values
```

```python
values = []
```

```python
values.append(draw_line(-7, 6, -10, 3))
values.append(draw_line(-10, 3, -6, 3))
values.append(draw_line(-6, 3, -11, -4))

values.append(draw_line(-4, 6, -1, 2))
values.append(draw_line(-1, 2, 1,5))
values.append(draw_line(1,5, 4,3))
values.append(draw_line(4,3, 6,8))

values.append(draw_line(6,10, 6,5))
values.append(draw_line(6,5, 6,-5))
values.append(draw_line(6,10, 10,6))
values.append(draw_line(6,5, 10,3))
values.append(draw_line(6,-5, 10,-4))
```

```python
x_values, y_values = zip(*values)
x_values = [item for sublist in x_values for item in (sublist + [None])]
y_values = [item for sublist in y_values for item in (sublist + [None])]
```

```python
def plot_func():
    plt.plot(x_values, y_values, color = 'black')
    # plt.plot(circleX, circleY, color = 'red')
    plt.title("Practice")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    plt.grid(True)
```

```python
plot_func()
```
